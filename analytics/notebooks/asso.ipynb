```python
import duckdb
import pandas as pd
import os
```

```python
df_asso1 = duckdb.read_csv("../data/raw/rna_waldec_20250901/rna_waldec_20250901_dpt_01.csv")
df_asso1
```

```python
duckdb.sql("""
    SELECT *
    FROM df_asso1
    WHERE adrs_codeinsee = 'A'
""")
```

```python
df_full=pd.DataFrame()
for files in os.listdir("../data/raw/rna_waldec_20250901/"):
    if files.endswith(".csv") and files.startswith("rna_waldec"):
        df_temp = pd.read_csv(
            os.path.join("../data/raw/rna_waldec_20250901/", files),
            engine="python"
           )
        df_temp = df_temp.loc[df_temp["position"] == "A"]
        df_temp = df_temp[["adrs_codeinsee","adrs_codepostal"]]
        df_full = pd.concat([df_full,df_temp], ignore_index=True,axis=0)
    
```

```python
df_asso = df_full.copy()
df_asso
```

```python
df_asso["adrs_codeinsee"] = df_asso["adrs_codeinsee"].astype(str)
df_asso["adrs_codepostal"] = df_asso["adrs_codepostal"].astype(str)
```

```python
cols = ["adrs_codeinsee", "adrs_codepostal"]
invalid_values = ["nan","<NA>", "NaN", "Nan", "0", "0.0", "", "INSEE","commune"]

df_asso[cols] = (
    df_asso[cols]
    .replace(invalid_values, pd.NA)
)
df_asso
```

```python
df_asso = df_asso.loc[
    ~df_asso[cols].isna().all(axis=1)
]

df_asso
```

```python
df_asso = df_asso.sort_values(["adrs_codeinsee","adrs_codepostal"])
df_asso
```

```python
df_asso["adrs_codepostal"] = df_asso["adrs_codepostal"].str.replace('.0','').str.zfill(5)
df_asso["adrs_codeinsee"] = df_asso["adrs_codeinsee"].str.replace('.0','').str.zfill(5)
df_asso.reset_index(drop=True, inplace=True)
df_asso
```

```python
df_nan = df_asso.loc[df_asso[["adrs_codeinsee","adrs_codepostal"]].isna().any(axis=1)]
df_nan
```

```python
print(df_nan.loc[df_nan["adrs_codeinsee"].isna() & df_nan["adrs_codepostal"].notna()].shape)
print(df_nan.loc[df_nan["adrs_codeinsee"].isna()].shape)
print(df_nan.loc[df_nan["adrs_codeinsee"].notna()].shape)
print(df_nan.loc[df_nan["adrs_codepostal"].isna()].shape)
print(df_nan.loc[df_nan["adrs_codepostal"].notna()].shape)
print(df_nan.loc[df_nan["adrs_codepostal"].isna() & df_nan["adrs_codeinsee"].notna()].shape)
```

```python
duckdb.sql(""" DESCRIBE df_com_explode""")
```

```python
query = """ 
SELECT DISTINCT e1.adrs_codeinsee, e2.code_insee, e2.code_postal
FROM df_nan_postal e1
LEFT JOIN df_com_explode e2
ON (e1.adrs_codeinsee = e2.code_insee)
ORDER BY e1.adrs_codeinsee
"""
df_sans_nan_postal=duckdb.sql(query).df().dropna()
df_sans_nan_postal
```

```python
df_sans_nan_postal = df_sans_nan_postal[["adrs_codeinsee","code_postal"]]
df_sans_nan_postal.rename(columns={"code_postal":"adrs_codepostal"}, inplace=True)
df_sans_nan_postal
```

```python
df_asso_complete = (pd.concat([df_asso[["adrs_codeinsee","adrs_codepostal"]],df_sans_nan_postal], ignore_index=True,axis=0)
                    .sort_values(["adrs_codeinsee","adrs_codepostal"])
                    .dropna())
df_asso_complete
```

```python
data_epci = duckdb.read_csv("../data/raw/perimetre-epci-a-fp.csv", ignore_errors=True, sep=";")
data_epci
```

```python
query = """
SELECT 
    e2.dept,
    e2.siren, 
    REPLACE(e2.total_pop_mun, ' ', '') AS population,
    count(*) AS nb_asso
FROM df_asso_complete e1
LEFT JOIN data_epci e2
ON e1.adrs_codeinsee = e2.insee
GROUP BY e2.dept,e2.siren,e2.total_pop_mun
ORDER BY dept, siren
"""

df_asso_epci = duckdb.sql(query).df().dropna()
```

```python
df_asso_epci
```

```python
q = """ 
SELECT *, round(1.0*TRY_CAST(nb_asso AS DOUBLE) / TRY_CAST(population AS DOUBLE) * 1000,2) as asso_per_1000_habitants
FROM df_asso_epci
ORDER BY dept,siren
"""

df_asso_summary = duckdb.query(q)
df_asso_summary.write_csv("../data/processed/asso_per_epci.csv")
```
